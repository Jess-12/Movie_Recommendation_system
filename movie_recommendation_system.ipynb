import dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data_movie= pd.read_csv('/content/movies.csv')
```

```python
data_movie.head()
```

```python
data_movie.shape
```

```python
#selecting relevant features 
selected_data=['genres','keywords','popularity','cast','director','tagline']
print(selected_data)
```

replacing all null values with null string

```python
for feature in selected_data:
  data_movie[feature]= data_movie[feature].fillna('')
```

```python
#combining selected features
combined_feature= data_movie['genres']+''+data_movie['keywords']+''+data_movie['tagline']+''+data_movie['title']+''+data_movie['cast']+''+data_movie['director']
print(combined_feature)
```

```python
vectorize=TfidfVectorizer()
vector_data=vectorize.fit_transform(combined_feature)
print(vector_data)
```

```python
#getting similarity score using cosine similarity
similarity= cosine_similarity(vector_data)
print(similarity.shape)
```

```python
movie_name= input("Enter the favourite movie name: ")
```

```python
#creating a list with all movie names
all_titles= data_movie['title'].tolist()
print(all_titles)
```

```python
# finding close match with the given user movie name
find_close_match= difflib.get_close_matches(movie_name,all_titles)
print(find_close_match)
```

```python
close_match= find_close_match[0]
print(close_match)
```

```python
movie_index = data_movie[data_movie.title==close_match]['index'].values[0]
print(movie_index)
```

```python
#getting list of similar movies
similarity_score= list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
#sorting based on similarity score
sorted_similar_movie= sorted(similarity_score,key = lambda x:x[1], reverse= True)
print(sorted_similar_movie)
```

```python
#printing name of similar mvoies
print('movies recommened : ' )
i=1
for movie in sorted_similar_movie:
  index= movie[0]
  title_from_movie= data_movie[data_movie.index==index]['title'].values[0]
  if i<20:
    print(i,'.',title_from_movie)
    i=i+1

```

MOVIE RECOMMENDATION SYSTEM:

```python
movie_name= input("Enter the favourite movie name: ")

all_titles= data_movie['title'].tolist()

find_close_match= difflib.get_close_matches(movie_name,all_titles)

close_match= find_close_match[0]

movie_index = data_movie[data_movie.title==close_match]['index'].values[0]

similarity_score= list(enumerate(similarity[movie_index]))

sorted_similar_movie= sorted(similarity_score,key = lambda x:x[1], reverse= True)

print('movies recommened : ' )
i=1
for movie in sorted_similar_movie:
  index= movie[0]
  title_from_movie= data_movie[data_movie.index==index]['title'].values[0]
  if i<20:
    print(i,'.',title_from_movie)
    i=i+1
```
